### 주요 해설 포인트

**인덱싱 관련**
- 파이썬 리스트는 0부터 시작
- 음수 인덱스는 뒤에서부터 계산 (-1이 마지막 요소)

**슬라이싱 관련**
- [시작:끝] - 끝 인덱스는 포함되지 않음
- [::단계] - 단계만큼 건너뛰며 선택

**메소드 관련**
- append(): 맨 뒤에 추가
- insert(위치, 값): 특정 위치에 삽입
- remove(값): 첫 번째로 나타나는 값 제거
- pop(인덱스): 해당 인덱스의 값을 제거하고 반환
