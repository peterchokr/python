### 주요 해설 포인트

**딕셔너리 기본 개념**
- 중괄호 `{}`를 사용하여 생성
- 키-값 쌍으로 구성: `{"key": "value"}`
- 키를 통한 빠른 데이터 접근

**키의 규칙**
- 불변(immutable) 타입만 가능: 문자열, 숫자, 튜플
- 리스트, 딕셔너리는 키로 사용 불가
- 키는 고유해야 함 (중복 불허)

**주요 메소드들**
- `keys()`: 모든 키 반환
- `values()`: 모든 값 반환  
- `items()`: 키-값 쌍 반환
- `get()`: 안전한 값 접근
- `pop()`: 항목 제거 후 값 반환
