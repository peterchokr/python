### 6번

````
class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed
````

````
class Vehicle:
    def __init__(self, brand):
        self.brand = brand
    
    def start(self):
        return f"{self.brand} 시동"

class Car(Vehicle):
    def __init__(self, brand, doors):
        super().__init__(brand)
        self.doors = doors
    
    def start(self):
        parent_start = super().start()
        return f"{parent_start} - {self.doors}개 문"

my_car = Car("현대", 4)
result = my_car.start()
print(result)
````

### 주요 해설 포인트

**상속 기본 개념**
- `class Child(Parent):` 문법으로 상속 구현
- 자식 클래스는 부모 클래스의 모든 속성과 메소드 상속
- IS-A 관계가 성립해야 올바른 상속 (학생은 사람이다)

**super() 함수**
- `super().method()` 형태로 부모 클래스 메소드 호출
- `__init__`에서 부모 클래스 초기화에 주로 사용
- 메소드 오버라이딩 시 부모 기능 확장에 활용

**메소드 오버라이딩**
