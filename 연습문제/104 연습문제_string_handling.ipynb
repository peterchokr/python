### 주요 해설 포인트

**문자열 기본 개념**
- 문자열은 따옴표로 둘러싸여야 함 (홑따옴표, 겹따옴표, 삼중따옴표)
- 인덱싱은 0부터 시작, 음수는 뒤에서부터
- 문자열은 불변(immutable) - 수정 불가, 새로운 문자열만 생성 가능

**인덱싱과 슬라이싱**
- 인덱싱: `string[index]` (단일 문자)
- 슬라이싱: `string[start:end]` (end는 포함되지 않음)
- 음수 인덱스: 뒤에서부터 계산 (-1이 마지막 문자)

**주요 메소드**
- 대소문자: `upper()`, `lower()`, `title()`, `capitalize()`
- 공백 처리: `strip()`, `lstrip()`, `rstrip()`
- 검색: `find()`, `count()`, `startswith()`, `endswith()`
- 검증: `isdigit()`, `isalpha()`, `isalnum()`

**문자열 형식화**
- f-문자열: `f"Hello {name}"` (권장)
- .format(): `"Hello {}".format(name)`
