### 주요 해설 포인트

**조건문 기본 문법**
- if 키워드와 콜론(:) 필수
- 들여쓰기로 코드 블록 구분
- elif는 이전 조건이 False일 때만 검사

**논리 연산자 이해**
- and: 모든 조건이 True여야 True
- or: 하나라도 True이면 True
- not: 조건을 반대로 변환

**조건문 실행 순서**
- if → elif → else 순서로 검사
- 조건이 True가 되면 해당 블록만 실행
- elif나 else는 이전 조건이 False일 때만 검사

**실용적 활용**
- 성적 관리, 나이별 분류, 할인 시스템
- 사용자 입력 검증과 오류 처리
- 복합 조건을 통한 정교한 판정

### 💡 성공 팁

**자주 하는 실수와 해결법**
- ❌ `if age > 18` → ✅ `if age > 18:`
- ❌ 잘못된 들여쓰기 → ✅ 일관된 4칸 들여쓰기
- ❌ `if x = 5` → ✅ `if x == 5`
- ❌ 논리 연산자 혼동 → ✅ and(그리고), or(또는) 명확히 구분

**조건문 설계 원칙**
- 가장 까다로운 조건부터 검사
- 상호 배타적인 조건들은 elif 사용
- 복잡한 조건은 변수로 분리
- 가독성을 위해 괄호 적극 활용

**디버깅 전략**
- 각 조건별로 print()로 확인
- 조건식의 결과를 직접 출력
- 경계값(boundary values)으로 테스트
- 예상치 못한 입력값으로 검증
