### 주요 해설 포인트

**중첩 리스트 접근**
- `list[행][열]` 형태로 접근
- 이중 인덱스 사용: `matrix[1][0]`
- 모든 요소 접근시 중첩 반복문 활용

**세트 연산**
- 교집합: `&` 또는 `intersection()`
- 합집합: `|` 또는 `union()`  
- 차집합: `-` 또는 `difference()`
- 빈 세트는 `set()`로 생성 (`{}`는 빈 딕셔너리)

**복합 데이터 구조**
- 딕셔너리 안의 리스트: `dict["key"][index]`
- 리스트 안의 딕셔너리: `list[index]["key"]`
- 적절한 데이터 구조 선택이 성능과 가독성에 중요
