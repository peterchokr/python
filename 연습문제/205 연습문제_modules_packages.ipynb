### 주요 해설 포인트

**Import 구문 문법**
- `import 모듈명`: 전체 모듈 가져오기
- `from 모듈명 import 함수명`: 특정 함수만 가져오기
- `import 모듈명 as 별칭`: 별칭으로 가져오기
- `from 모듈명 import *`: 모든 요소 가져오기 (권장하지 않음)

**표준 라이브러리 모듈**
- math: 수학 계산 (sqrt, pi, ceil, floor 등)
- random: 난수 생성 (randint, choice, shuffle 등)
- datetime: 날짜/시간 처리 (now, today, strftime 등)

**모듈 사용의 장점**
- 코드 재사용성 향상
- 코드 구조화 및 관리 용이
- 네임스페이스 분리로 이름 충돌 방지
- 팀 협업 시 역할 분담 가능
