- 대소문자 구분 (`name`과 `Name`은 다른 변수)

**데이터 타입**
- **int**: 정수 (25, -10, 0)
- **float**: 실수 (3.14, 175.5)  
- **str**: 문자열 ("안녕", '김철수')
- **bool**: 불린 (True, False - 대문자 주의)

**type() 함수**
- 변수의 데이터 타입 확인
- `<class 'int'>` 형태로 출력
- 디버깅과 데이터 검증에 활용
