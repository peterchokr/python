### 주요 해설 포인트

**파일 기본 개념**
- open() 함수로 파일 열기
- 파일 모드: "r"(읽기), "w"(쓰기/덮어쓰기), "a"(추가)
- 한글 처리시 encoding="utf-8" 필수

**파일 읽기 메소드**
- read(): 전체 내용을 문자열로
- readline(): 한 줄씩 문자열로
- readlines(): 모든 줄을 리스트로

**with 문의 장점**
- 자동 파일 닫기
- 예외 발생 시에도 안전한 처리
- 코드 간소화
