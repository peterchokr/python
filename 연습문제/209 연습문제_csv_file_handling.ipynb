### 주요 해설 포인트

**CSV 기본 개념**
- CSV = Comma-Separated Values (쉼표로 구분된 값)
- 범용성, 단순함, 가벼움이 주요 장점
- 보안성은 장점이 아님 (평문 저장)

**파일 읽기/쓰기**
- 읽기: `'r'` 모드, 쓰기: `'w'` 모드
- 한글 처리: `encoding='utf-8'`
- 쓰기 시: `newline=''` 파라미터 필수

**csv 모듈 사용법**
- `csv.reader`: 리스트 형태로 읽기
- `csv.DictReader`: 딕셔너리 형태로 읽기 (열 이름 접근 가능)
- `csv.writer`/`csv.DictWriter`: 각각 대응하는 쓰기 방법

**데이터 처리**
- 대용량 파일은 한 줄씩 처리하여 메모리 효율성 확보
- 누락 데이터는 조건문으로 확인 후 적절히 처리
- 데이터 타입 변환 시 예외 처리 필요
