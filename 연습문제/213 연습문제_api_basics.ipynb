### 20번
다음 코드의 실행 과정을 단계별로 분석하고, 각 부분의 역할과 중요성을 상세히 설명하세요.

````
import requests
import json

def get_weather_data(city, api_key):
    url = f"https://api.weather.com/current?city={city}"
    headers = {
        'Authorization': f'Bearer {api_key}',
        'Content-Type': 'application/json'
    }
    
    try:
        response = requests.get(url, headers=headers, timeout=10)
        if response.status_code == 200:
            weather_data = response.json()
            return weather_data
        else:
            print(f"API Error: {response.status_code}")
            return None
    except Exception as e:
        print(f"Request failed: {e}")
        return None
````

### 주요 해설 포인트

**API 기본 개념**
- API는 서로 다른 소프트웨어 간 데이터 교환을 위한 규칙과 도구
- RESTful API가 현재 웹에서 가장 널리 사용되는 방식
- HTTP 메서드별 용도: GET(조회), POST(생성), PUT(수정), DELETE(삭제)

**JSON 데이터 처리**
- JSON은 JavaScript Object Notation의 줄임말
- `json.loads()`: JSON 문자열 → 파이썬 객체
- `json.dumps()`: 파이썬 객체 → JSON 문자열
- 문자열은 반드시 큰따옴표로 감싸야 함

**HTTP 상태 코드**
- 200: 성공 (OK)
- 400: 잘못된 요청 (Bad Request)  
- 401: 인증 실패 (Unauthorized)
- 403: 접근 금지 (Forbidden)
- 404: 리소스 없음 (Not Found)
- 500: 서버 오류 (Internal Server Error)

**API 보안 및 인증**
- API 키는 환경 변수나 설정 파일로 관리
- Authorization 헤더를 통한 인증
- Rate Limiting으로 과도한 요청 방지
