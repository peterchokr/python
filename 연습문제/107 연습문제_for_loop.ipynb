### 주요 해설 포인트

**for문 기본 문법**
- for 키워드와 콜론(:) 필수
- 들여쓰기로 반복 블록 구분
- range() 함수로 숫자 시퀀스 생성

**range() 함수 이해**
- range(n): 0부터 n-1까지
- range(start, stop): start부터 stop-1까지
- range(start, stop, step): 증가값 지정 가능

**문자열 반복**
- 문자열의 각 문자에 순차적 접근
- 문자 분석과 처리에 유용
- len() 함수와 조합하여 인덱스 활용

**중첩 반복문**
- 패턴 생성과 표 만들기에 활용
- 바깥 반복 × 안쪽 반복 = 총 실행 횟수
- 들여쓰기 수준으로 중첩 깊이 구분
