- 프로그래밍 언어는 인간과 컴퓨터 간의 소통 도구

**파이썬의 특징**
- 인터프리터 언어: 컴파일 없이 바로 실행
- 읽기 쉬운 문법: 영어와 유사한 구조
- 크로스 플랫폼: 다양한 운영체제에서 실행 가능

**설치 및 환경설정**
- "Add Python to PATH" 체크 필수
- `python --version` 명령어로 설치 확인
- 확장자 `.py` 사용

**개발환경 비교**
- IDLE: 초보자용, 기본 기능
- VS Code: 중급자용, 확장 기능 풍부
- PyCharm: 전문가용, 고급 기능
