### 주요 해설 포인트

**데이터 분석 기본 개념**
- 데이터 분석의 목적: 패턴 찾기, 예측, 의사결정 지원
- 데이터 크기 줄이기는 데이터 압축의 개념이지 분석의 목적이 아님
- 객관성이 핵심이므로 주관성은 특징이 아님

**기본 통계량**
- 평균 = sum(data) / len(data)
- 최댓값, 최솟값: max(), min() 함수 사용
- 범위 = 최댓값 - 최솟값

**데이터 필터링**
- 조건문(if)을 사용하여 특정 조건 만족 데이터 선택
- and, or 연산자로 복합 조건 생성
- 파이썬에서는 &&가 아닌 and 사용

**데이터 그룹화**
