### 주요 해설 포인트

**Matplotlib 기본 개념**
- Matplotlib은 파이썬의 대표적인 데이터 시각화 라이브러리
- `import matplotlib.pyplot as plt`가 표준 import 방법
- `plt.show()`로 그래프를 화면에 표시

**그래프 유형별 특징**
- **선 그래프**: 시계열 데이터, 트렌드 분석에 적합
- **막대 그래프**: 카테고리간 비교, 이산적 데이터에 적합
- **히스토그램**: 연속형 데이터의 분포 확인에 적합
- **원형 그래프**: 전체 대비 비율 표시에 적합

**그래프 구성 요소**
- 제목: `plt.title()`
- 축 라벨: `plt.xlabel()`, `plt.ylabel()`
- 범례: `plt.legend()`
- 격자: `plt.grid()`

**커스터마이징 옵션**
- 색상, 마커, 선 스타일 지정 가능
- `figsize`로 그래프 크기 조절
- 서브플롯으로 여러 그래프 배치
