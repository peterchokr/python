### 주요 해설 포인트

**리스트 컴프리헨션 관련**
- 기본 구조: `[표현식 for 항목 in 반복가능객체]`
- 조건 포함: `[표현식 for 항목 in 반복가능객체 if 조건]`
- 복잡한 로직은 일반 반복문이 더 적합

**튜플 관련**
- 불변(immutable) 특성으로 수정 불가
- 단일 항목 튜플은 쉼표 필수: `(42,)`
- 언패킹 시 변수 개수와 요소 개수가 일치해야 함

**고급 함수들**
- enumerate(): 인덱스와 값을 함께 반환
- zip(): 여러 시퀀스를 병렬로 순회
- 코드의 가독성과 효율성을 높여줌
