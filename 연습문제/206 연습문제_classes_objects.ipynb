### 주요 해설 포인트

**클래스와 객체 기본 개념**
- 클래스는 설계도, 객체는 실제 인스턴스
- `class` 키워드로 클래스 정의
- 클래스명은 대문자로 시작하는 관례

**메소드와 속성**
- 속성: 클래스 내의 변수 (데이터 저장)
- 메소드: 클래스 내의 함수 (동작 수행)
- `self` 매개변수로 현재 객체 참조

**__init__ 메소드**
- 객체 생성 시 자동 호출
- 객체의 초기 상태 설정
- `self.속성명 = 값` 형태로 속성 초기화
