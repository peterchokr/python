### 19번

````
def test_function(x, y):
    try:
        result = x / y
        print(f"결과: {result}")
        return result
    except ZeroDivisionError:
        print("0으로 나눌 수 없습니다")
        return None
    finally:
        print("함수 실행 완료")

test_function(10, 2)  # 경우 1
test_function(10, 0)  # 경우 2
````

### 주요 해설 포인트

**예외 유형별 특징**
- ZeroDivisionError: 숫자를 0으로 나눌 때
- ValueError: 적절하지 않은 값으로 변환 시도
- TypeError: 호환되지 않는 타입 간 연산
- IndexError: 리스트 범위 초과 접근
- KeyError: 딕셔너리에 없는 키 접근
- FileNotFoundError: 존재하지 않는 파일 접근

**try-except 구조**
- try: 예외가 발생할 수 있는 코드
- except: 예외 처리 코드
- else: 예외가 발생하지 않았을 때 실행
- finally: 항상 실행되는 정리 코드

**finally 블록의 특성**
- return문이 있어도 finally는 실행됨
- 예외 발생 여부와 관계없이 항상 실행
- 주로 자원 정리 용도로 사용
