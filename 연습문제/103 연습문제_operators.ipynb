### 주요 해설 포인트

**산술 연산자**
- `/` : 실수 나눗셈, 항상 float 결과
- `//` : 정수 나눗셈, 몫만 반환
- `%` : 나머지 연산자
- `**` : 거듭제곱 연산자

**비교 연산자**
- 항상 `True` 또는 `False` 반환
- `==` (비교)와 `=` (할당) 구분 중요
- 연쇄 비교 가능: `1 < x < 10`

**논리 연산자**
- `and`: 모든 조건이 참
- `or`: 하나라도 참
- `not`: 조건을 반대로
- 단축 평가(Short-circuit) 적용

**연산자 우선순위**
1. `()` (괄호)
2. `**` (거듭제곱)
3. `*`, `/`, `//`, `%` (곱셈과 나눗셈)
4. `+`, `-` (덧셈과 뺄셈)
5. 비교 연산자
6. 논리 연산자
