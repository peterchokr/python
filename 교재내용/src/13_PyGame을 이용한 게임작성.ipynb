p.463 우주선 움직이기

```python
import pygame

pygame.init()
display = pygame.display.set_mode((800, 600))

player=pygame.image.load("spaceship.png")
playerX, playerY, playerDx , playerDy = 400, 550, 0, 0


alien=pygame.image.load("alien.png")    # 외계 우주선 이미지를 읽는다.
alienX, alienY, alienDx , alienDy   = 0, 10, 0.1, 0.1

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        if event.type == pygame.KEYDOWN:    # 키가 눌러지면 
            if event.key == pygame.K_LEFT:  # 왼쪽 화살표 키, 오른쪽 화살표 키에 따라 위치 변경
                playerDx = -0.1
            if event.key == pygame.K_RIGHT:             
                playerDx = 0.1
        if event.type == pygame.KEYUP:  # 방향키에서 손을 떼면 멈추도록 이동 크기를 0으로.
            if event.key == pygame.K_RIGHT or event.key == pygame.K_LEFT:
                playerDx = 0

    playerX += playerDx     # 우주선 움직이기(이동)
    
    # 이미지가 화면을 벗어나는 문제를 해결
    # if playerX <= 0:
    #     playerX = 0
    # if playerX >= 750:
    #     playerX = 750

    display.fill((0, 0, 0))
    display.blit(player, (playerX, playerY))
    display.blit(alien, (alienX, alienY))   # 외계 우주선 이미지를 화면에 표시
    pygame.display.update()

pygame.quit()
```

p.466 외계 우주선을 움직여 보자

```python
import pygame

pygame.init()
display = pygame.display.set_mode((800, 600))

player=pygame.image.load("spaceship.png")
playerX, playerY, playerDx , playerDy = 400, 550, 0, 0


alien=pygame.image.load("alien.png")    # 외계 우주선 이미지를 읽는다.
alienX, alienY, alienDx , alienDy   = 0, 10, 0.1, 0.1

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        if event.type == pygame.KEYDOWN:    
            if event.key == pygame.K_LEFT:  
                playerDx = -0.1
            if event.key == pygame.K_RIGHT:             
                playerDx = 0.1
        if event.type == pygame.KEYUP:  
            if event.key == pygame.K_RIGHT or event.key == pygame.K_LEFT:
                playerDx = 0

    playerX += playerDx     
    
    # 이미지가 화면을 벗어나는 문제를 해결
    # if playerX <= 0:
    #     playerX = 0
    # if playerX >= 750:
    #     playerX = 750

    # 외계 우주선을 x방향으로 이동. 
    # 0보다 작거나 750보다 크면 y좌표에 30이 더해져 아래쪽으로 내려오게 한다.
    alienX += alienDx
    if  alienX <= 0 or alienX > 750:
        alienDx *= -1
        alienY += 30

    display.fill((0, 0, 0))
    display.blit(player, (playerX, playerY))
    display.blit(alien, (alienX, alienY))   
    pygame.display.update()

pygame.quit()
```

p.468 미사일 만들기

```python
import pygame

pygame.init()
display = pygame.display.set_mode((800, 600))

player=pygame.image.load("spaceship.png")
playerX, playerY, playerDx , playerDy = 400, 550, 0, 0

alien=pygame.image.load("alien.png")
alienX, alienY, alienDx , alienDy   = 0, 10, 0.1, 0.1

missile = pygame.image.load('missile.png')      # 미사일 이미지를 읽는다.
missileX, missileY, missileDx , missileDy   = 0, 1000, 0, 0.1

missileState = "hidden"

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:                playerDx = -0.1
            if event.key == pygame.K_RIGHT:             playerDx = 0.1
            
            # 스페이스바를 누르면 미사일을 발사한다.
            # 미사일 상태가 'hidden'이면 'fire'로 바꾸고 미사일 위치를 우주선 위치로 설정한다.
            if event.key == pygame.K_SPACE:
                if missileState == "hidden":
                    missileState = "fire"
                    missileX, missileY = playerX, playerY
                    
        if event.type == pygame.KEYUP:
            if event.key == pygame.K_RIGHT or event.key == pygame.K_LEFT:
                playerDx = 0

    playerX += playerDx

    alienX += alienDx
    if  alienX <= 0 or alienX > 750:
        alienDx *= -1
        alienY += 30

    # 미사일 y 좌표가 0보다 작으면 미사일이 화면을 벗어난 것
    # 미사일을 화면에 숨기고 상태를 'hidden'으로 변경
    if missileY <= 0:       
        missileY = 1000
        missileState = "hidden"

    # 미사일 상태가 'fire'이면 미사일 y 좌표를 missileDy 만큼 감소시켜 위쪽으로 이동하게 한다.
    if missileState == "fire":
        missileY -= missileDy

    display.fill((0, 0, 0))
    display.blit(player, (playerX, playerY))

    display.blit(missile, (missileX, missileY)) # 미사일 화면을 그린다.
    display.blit(alien, (alienX, alienY))
    pygame.display.update()

pygame.quit()
```

p.469 충돌 감지

```python
import pygame

pygame.init()
display = pygame.display.set_mode((800, 600))

# 점수를 표시하기 위해 변수 score 생성
myfont = pygame.font.SysFont('Comic Sans MS', 30)
score = 0

player=pygame.image.load("spaceship.png")
playerX, playerY, playerDx , playerDy = 400, 550, 0, 0

alien=pygame.image.load("alien.png")
alienX, alienY, alienDx , alienDy   = 0, 10, 0.1, 0.1

missile = pygame.image.load('missile.png')
missileX, missileY, missileDx , missileDy   = 0, 1000, 0, 0.1
missileState = "hidden"

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:              playerDx = -0.1
            if event.key == pygame.K_RIGHT:             playerDx = 0.1
            if event.key == pygame.K_SPACE:
                if missileState == "hidden":
                    missileState = "fire"
                    missileX, missileY =playerX, playerY
        if event.type == pygame.KEYUP:
            if event.key == pygame.K_RIGHT or event.key == pygame.K_LEFT:
                playerDx = 0

    # 각 이미지를 둘러싸고 있는 사각형을 얻어서 겹치는지를 colliderect()로 검사
    # 충돌이 감지되면 점수를 1 증가시키고, 외계선의 위치를 초기화
    rect1 = pygame.Rect(alien.get_rect(topleft=(alienX, alienY)))
    rect2 = pygame.Rect(missile.get_rect(topleft=(missileX, missileY)))
    if rect1.colliderect(rect2) and missileState != "hidden":
        score += 1
        alienX, alienY, alienDx , alienDy   = 0, 10, 0.1, 0.1
        
        # 부딪힌 후 미사일이 사라지도록 하기 위하여
        # missileX, missileY, missileDx , missileDy   = alienX, 1000, 0, 0.1
        # missileState = "hidden"
        
    playerX += playerDx

    alienX += alienDx
    if  alienX <= 0 or alienX > 750:
        alienDx *= -1
        alienY += 30

    if missileY <= 0:
        missileY = 1000
        missileState = "hidden"

    if missileState == "fire":
        missileY -= missileDy

    display.fill((0, 0, 0))
    display.blit(player, (playerX, playerY))
    display.blit(missile, (missileX, missileY))
    display.blit(alien, (alienX, alienY))
    
    # 점수를 나타내는 score를 (10, 550) 위치에 흰색으로 표시
    text = myfont.render(f'score={score}', False, (255, 255, 255))
    display.blit(text,(10,550))
    
    pygame.display.update()

pygame.quit()
```

p.472 외계 우주선 많이 생성하기

```python
import pygame

pygame.init()
display = pygame.display.set_mode((800, 600))
myfont = pygame.font.SysFont('Comic Sans MS', 30)
score = 0

player = pygame.image.load("spaceship.png")
playerX, playerY, playerDx , playerDy = 400, 550, 0, 0

# -------
# 여러개 만들어 지는 외계 우주선의 x좌표, y좌표, 이동속도를 리스트에 저장
alien=pygame.image.load("alien.png")
alienX = [ ]
alienY = [ ]
alienDx = [ ]
alienDy = [ ]
alienNumber = 6

for i in range(alienNumber):
    alienX.append(20+i*60)
    alienY.append(10)
    alienDx.append(0.1)
    alienDy.append(0.0)
# ------    

missile = pygame.image.load('missile.png')
missileX, missileY, missileDx , missileDy   = 0, 1000, 0, 0.1
missileState = "hidden"

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:              playerDx = -0.1
            if event.key == pygame.K_RIGHT:             playerDx = 0.1
            if event.key == pygame.K_SPACE:
                if missileState == "hidden":
                    missileState = "fire"
                    missileX, missileY =playerX, playerY
        if event.type == pygame.KEYUP:
            if event.key == pygame.K_RIGHT or event.key == pygame.K_LEFT:
                playerDx = 0

    playerX += playerDx

    # ------
    # 리스트에서 외계 우주선을 하나씩 꺼내서 처리
    display.fill((0, 0, 0))
    for i in range(alienNumber):
        alienX[i] += alienDx[i] # 위치를 업데이트
        alienY[i] += alienDy[i]
        if  alienX[i] <= 0 or alienX[i] > 750:  # 화면을 벗어나면 움직이는 방향을 바꾼다.
            alienDx[i] *= -1
            alienY[i] += 30
        rect1 = pygame.Rect(alien.get_rect(topleft=(alienX[i], alienY[i]))) # 충동감지 코드
        rect2 = pygame.Rect(missile.get_rect(topleft=(missileX, missileY)))
        if rect1.colliderect(rect2) and missileState != "hidden":
            score += 1
            alienX[i], alienY[i], alienDx[i] , alienDy[i] = 0, 1000, 0.1, 0.0
        display.blit(alien, (alienX[i], alienY[i])) # 외계 우주선을 화면에 그린다.
    # -------

    if missileY <= 0:
        missileY = 1000
        missileState = "hidden"

    if missileState == "fire":
        missileY -= missileDy

    display.blit(player, (playerX, playerY))
    display.blit(missile, (missileX, missileY))
    text = myfont.render(f'score={score}', False, (255, 255, 255))

    display.blit(text,(10,550))
    pygame.display.update()

pygame.quit()
```
